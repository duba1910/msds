Here's a table showing the appropriate symbols for the population and sample of various descriptive statistics:

| **Statistic**              | **Population Symbol** | **Sample Symbol** |
|----------------------------|-----------------------|-------------------|
| **Mean**                   |  (mu)                | \(\overline{x}\)  |
| **Median**                 | M or  (eta)          | \( \tilde{x} \)   |
| **Mode**                   | Mo                    | Mo                |
| **Range**                  | R                     | R                 |
| **Interquartile Range**    | IQR                   | IQR               |
| **Variance**               |  (sigma squared)    | s                |
| **Standard Deviation**     |  (sigma)             | s                 |
| **Skewness**               | 3 (gamma 1)          | G3 or \( \tilde{g}_3 \) |
| **Kurtosis**               | 4 (gamma 4)          | G4 or \( \tilde{g}_4 \) |
| **Coefficient of Correlation** |  (rho)          | r                 |

### Notes:
- **Mean:** The population mean is denoted by , while the sample mean is \(\overline{x}\).
- **Median and Mode:** While the population and sample symbols are generally the same, specific contexts might have variations.
- **Variance and Standard Deviation:** The population variance is denoted by  and standard deviation by , while the sample variance and standard deviation are denoted by s and s, respectively.
- **Skewness and Kurtosis:** For population skewness and kurtosis,  and  are used, while the sample counterparts are often denoted by G, G, or by using the sample estimate symbols \( \tilde{g}_1 \) and \( \tilde{g}_2 \).
- **Coefficient of Correlation:** The population correlation coefficient is denoted by  (rho), while the sample correlation coefficient is denoted by r.

These symbols are standard, but different textbooks or contexts may have slight variations.
